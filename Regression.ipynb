```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv(r"C:\Users\shita\Downloads\dataset.csv")
```

```python
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,1].values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
X_train,X_test,y_train,y_test
```

```python
from sklearn.linear_model import LinearRegression
reg= LinearRegression()
reg.fit(X_train,y_train)
```

```python
y_predict=reg.predict(X_test)
y_predict
```

```python
y_test
```

```python
plt.scatter(X_train,y_train,color='red')
plt.plot(X_train,reg.predict(X_train),color='blue')
plt.title("Linear Regression salaray vs experiances",size=18)
plt.xlabel("Year of Employee",size=15)
plt.ylabel("Saleries of Employee",size=15)
plt.show()
```

```python
plt.scatter(X_test,y_test,color='red')
plt.plot(X_test,reg.predict(X_test),color='Blue')
plt.title("linear regression salary vs experince", size=16)
plt.xlabel("Year of Employee", size=12)
plt.ylabel("Saleries of Employee", size=12)
plt.show()
```
